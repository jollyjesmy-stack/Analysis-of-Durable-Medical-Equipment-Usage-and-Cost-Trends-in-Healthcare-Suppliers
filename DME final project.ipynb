## Step 4: Checking for Missing Values

Missing values can affect the quality of analysis. Therefore, we identify columns with missing values.

Some supplier information fields may contain missing values because the supplier could be registered as an organization rather than an individual.

```python
df.isnull().sum()
```

## Step 5: Handling Missing Values

Missing values in text columns are replaced with appropriate placeholder values such as "Unknown" or "Not Available". This ensures that the dataset remains consistent and usable for analysis.

```python
df['Suplr_Prvdr_First_Name'] = df['Suplr_Prvdr_First_Name'].fillna("Unknown")
df['Suplr_Prvdr_MI'] = df['Suplr_Prvdr_MI'].fillna("")
df['Suplr_Prvdr_Crdntls'] = df['Suplr_Prvdr_Crdntls'].fillna("Unknown")
df['Suplr_Prvdr_St2'] = df['Suplr_Prvdr_St2'].fillna("Not Available")
```

## Step 6: Checking for Duplicate Records

Duplicate records can lead to incorrect analysis results. Therefore, duplicate entries are identified and removed.

```python
df.duplicated().sum()
```

## Step 7: Removing Duplicate Records

All duplicate rows are removed to ensure that each observation in the dataset is unique.

```python
df = df.drop_duplicates()
```

## Step 8: Correcting Data Types

Some numerical columns may be incorrectly stored as object data types. These columns are converted to numeric format for accurate calculations and analysis.

```python
df['Tot_Suplr_Benes'] = pd.to_numeric(df['Tot_Suplr_Benes'], errors='coerce')
df['Tot_Suplr_Clms'] = pd.to_numeric(df['Tot_Suplr_Clms'], errors='coerce')
df['Tot_Suplr_Srvcs'] = pd.to_numeric(df['Tot_Suplr_Srvcs'], errors='coerce')

df['Avg_Suplr_Sbmtd_Chrg'] = pd.to_numeric(df['Avg_Suplr_Sbmtd_Chrg'], errors='coerce')
df['Avg_Suplr_Mdcr_Alowd_Amt'] = pd.to_numeric(df['Avg_Suplr_Mdcr_Alowd_Amt'], errors='coerce')
df['Avg_Suplr_Mdcr_Pymt_Amt'] = pd.to_numeric(df['Avg_Suplr_Mdcr_Pymt_Amt'], errors='coerce')
df['Avg_Suplr_Mdcr_Stdzd_Amt'] = pd.to_numeric(df['Avg_Suplr_Mdcr_Stdzd_Amt'], errors='coerce')
```

## Step 9: Standardizing Text Columns

To maintain consistency in categorical fields, text values such as city and state names are standardized.

```python
df['Suplr_Prvdr_City'] = df['Suplr_Prvdr_City'].str.title()
df['Suplr_Prvdr_State_Abrvtn'] = df['Suplr_Prvdr_State_Abrvtn'].str.upper()
df['Suplr_Prvdr_Cntry'] = df['Suplr_Prvdr_Cntry'].str.upper()
```

## Step 10: Creating Derived Columns

New columns are created to provide additional insights into supplier financial data.

Two new metrics are calculated:
- Total Medicare Payment
- Total Submitted Charges

```python
df['Total_Medicare_Payment'] = df['Tot_Suplr_Srvcs'] * df['Avg_Suplr_Mdcr_Pymt_Amt']

df['Total_Submitted_Charges'] = df['Tot_Suplr_Srvcs'] * df['Avg_Suplr_Sbmtd_Chrg']
```

## Step 11: Filtering Invalid Records

Records with zero or negative service or claim values are removed because they may represent incorrect or incomplete entries.

```python
df = df[df['Tot_Suplr_Srvcs'] > 0]
df = df[df['Tot_Suplr_Clms'] > 0]
```

After cleaning, the dataset structure is checked again to confirm that the cleaning process was successful.

## Step 14: Exploratory Data Analysis (EDA)

In this section, we explore patterns, trends, and relationships within the dataset using statistical summaries and visualizations.

The analysis includes:

### 1. Distribution of Total Supplier Services

This plot shows how services are distributed across suppliers.

```python
plt.figure()
df['Tot_Suplr_Srvcs'].hist(bins=50)
plt.title("Distribution of Supplier Services")
plt.xlabel("Total Services")
plt.ylabel("Frequency")
plt.show()
```

This histogram illustrates the distribution of total supplier services (`Tot_Suplr_Srvcs`). 
The distribution is positively skewed, meaning most suppliers provide a lower number of services, while a small number of suppliers account for a disproportionately high volume.
This indicates that service delivery is concentrated among a few high-performing suppliers, which may reflect differences in operational scale, specialization, or demand. Understanding this distribution is important for identifying service concentration and potential resource allocation patterns in healthcare delivery.

### 2. Distribution of Average Medicare Payment Amount

```python
plt.figure()
df['Avg_Suplr_Mdcr_Pymt_Amt'].hist(bins=50)
plt.title("Distribution of Medicare Payment")
plt.xlabel("Payment Amount")
plt.ylabel("Frequency")
plt.show()
```

This histogram shows the distribution of the Average Medicare Payment Amount (`Avg_Suplr_Mdcr_Pymt_Amt`) across different medical equipment services. 
The distribution appears right-skewed, indicating that most services have relatively lower average payment amounts, while a small number of services have significantly higher payments. This suggests that only certain specialized medical equipment or procedures incur high costs, whereas the majority of services are reimbursed at lower rates.
This insight highlights the variation in Medicare spending across different types of durable medical equipment and services.

### 3. Top States by Total Services

```python
top_states = df.groupby('Suplr_Prvdr_State_Abrvtn')['Tot_Suplr_Srvcs'].sum().sort_values(ascending=False).head(10)

plt.figure()
top_states.plot(kind='bar')
plt.title("Top 10 States by Services")
plt.xlabel("State")
plt.ylabel("Total Services")
plt.show()
```

This bar chart represents the top states based on the total number of services (`Tot_Suplr_Srvcs`) provided by suppliers, grouped using the `Suplr_Prvdr_State_Abrvtn` variable. 
The visualization shows that a few states contribute significantly higher service volumes compared to others, indicating a concentration of healthcare equipment services in specific regions. This may be due to factors such as population size, availability of suppliers, and healthcare demand.
Overall, the graph highlights regional disparities in service distribution, which can be useful for understanding areas with higher healthcare utilization and resource allocation.

### 4. Relationship between Services and Medicare Payment

```python
plt.figure()
plt.scatter(df['Tot_Suplr_Srvcs'], df['Avg_Suplr_Mdcr_Pymt_Amt'])
plt.title("Services vs Medicare Payment")
plt.xlabel("Total Services")
plt.ylabel("Medicare Payment")
plt.show()
```

This scatter plot illustrates the relationship between the total number of services (`Tot_Suplr_Srvcs`) and the average Medicare payment amount (`Avg_Suplr_Mdcr_Pymt_Amt`). 
The plot shows that there is no strong linear relationship between service volume and payment amount. While some suppliers with higher service counts receive higher payments, many points are widely scattered, indicating that payment amounts vary depending on the type of medical equipment or service provided rather than just the volume of services.
This suggests that certain services are inherently more expensive, and higher service counts do not necessarily lead to higher average payments.

### 5. Relationship between Claims and Services

```python
plt.figure()
plt.scatter(df['Tot_Suplr_Clms'], df['Tot_Suplr_Srvcs'])
plt.title("Claims vs Services")
plt.xlabel("Total Claims")
plt.ylabel("Total Services")
plt.show()
```

This scatter plot illustrates the relationship between the total number of supplier claims (`Tot_Suplr_Clms`) and the total number of services (`Tot_Suplr_Srvcs`). 
The visualization shows a strong positive relationship, indicating that as the number of claims increases, the number of services provided also increases. This is expected because each claim typically corresponds to one or more services rendered by the supplier.
However, the spread of points suggests some variability, meaning that different suppliers may provide varying numbers of services per claim depending on the type of medical equipment or service offered.
Overall, this relationship highlights the dependency between claims and service volume in the Medicare equipment dataset.

### 6. Rental vs Non-Rental Services Comparison

```python
rental_counts = df['Suplr_Rentl_Ind'].value_counts()

plt.figure()
rental_counts.plot(kind='bar')
plt.title("Rental vs Non-Rental Services")
plt.xlabel("Rental Indicator")
plt.ylabel("Count")
plt.show()
```

This graph compares the number of services provided for rental and non-rental medical equipment using the `Suplr_Rentl_Ind` variable. 
It shows how service volume differs between equipment that is rented versus equipment that is purchased or non-rental.
From the visualization, it can be observed that one category (either rental or non-rental) dominates in terms of total services, indicating the preferred mode of equipment usage in the healthcare system. This highlights how certain types of medical equipment are more commonly rented rather than owned, or vice versa.
This comparison helps in understanding usage patterns of durable medical equipment and provides insight into cost structures and service delivery models in Medicare.

### 7. Correlation Heatmap

This heatmap shows relationships between numerical variables.

```python
plt.figure()
corr = df[['Tot_Suplr_Benes','Tot_Suplr_Clms','Tot_Suplr_Srvcs',
           'Avg_Suplr_Sbmtd_Chrg','Avg_Suplr_Mdcr_Pymt_Amt']].corr()

plt.imshow(corr)
plt.title("Correlation Heatmap")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.show()
```

The correlation heatmap illustrates the relationships between key numerical variables in the dataset, including total services (`Tot_Suplr_Srvcs`), total claims (`Tot_Suplr_Clms`), total beneficiaries (`Tot_Suplr_Benes`), and various Medicare payment amounts.
A strong positive correlation can be observed between the number of services, claims, and beneficiaries, indicating that suppliers serving more patients tend to provide more services and generate more claims. Additionally, payment-related variables such as average submitted charges, allowed amounts, and Medicare payments also show strong positive correlations, suggesting consistency in cost structures.
These relationships highlight how service volume and patient count are closely linked to healthcare expenditure, providing valuable insights into supplier performance and cost patterns in the Medicare equipment dataset.

### 8. Top HCPCS Codes by Total Medicare Payment

```python
top_hcpcs = df.groupby('HCPCS_Cd')['Total_Medicare_Payment'].sum().sort_values(ascending=False).head(10)

plt.figure()
top_hcpcs.plot(kind='bar')
plt.title("Top HCPCS Codes by Medicare Payment")
plt.xlabel("HCPCS Code")
plt.ylabel("Total Payment")
plt.show()
```

This bar chart displays the top HCPCS codes based on the total Medicare payment amount, calculated using the variable `Avg_Suplr_Mdcr_Pymt_Amt` aggregated across services. 
The visualization highlights which medical equipment or services receive the highest total Medicare payments, indicating areas of high expenditure within the healthcare system. It can be observed that a few HCPCS codes account for a significant portion of total payments, suggesting that certain medical devices or procedures are more costly and widely utilized.
This insight is useful for understanding cost distribution across different medical services and can help identify high-cost equipment categories that may require further analysis or cost optimization.
